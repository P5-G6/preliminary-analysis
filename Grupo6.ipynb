Dataset escolhido: [A cachaça de Alambique, um estudo sobre hábitos de consumo em goiânia.](https://www.sebrae.com.br/Sebrae/Portal%20Sebrae/UFs/GO/Sebrae%20de%20A%20a%20Z/A%20Cacha%C3%A7a%20de%20Alambique%20-%20Um%20estudo%20sobre%20o%20h%C3%A1bito%20de%20Consumoi%20em%20Goi%C3%A2nia.pdf)



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Evoluções do comércio de cachaça entre 2009-2017

```python
columns = [
    'Produção (kL)',
    'Participação das Produção (%)',
    'Vendas (kL)',
    'Participação das Vendas (%)',
]

data = [
    [1176404, 8.1, 989464, 7.4], 
    [1047997, 6.9, 942770, 6.5],
    [1252133, 7.9, 1063716, 7.3],
    [1147899, 7.1, 930732, 6.1],
    [978649, 6.4, 868000, 6.4],
    [884302, 5.5, 777311, 5.7],
    [931621, 5.8, 777989, 5.6],
    [1035967, 6.5, 897172, 6.5],
    [970538, 6.2, 810611, 6.4],
]

years = range(2009, 2018)

df_evolution = pd.DataFrame(
    np.array(data),
    columns=columns,
    index=years,
)
sns.lineplot(data=df_evolution[['Produção (kL)', 'Vendas (kL)']])
plt.show()

sns.lineplot(data=df_evolution[['Participação das Produção (%)', 'Participação das Vendas (%)']])
plt.show()
```

## Relação entre exportação e produtores de Cachaça

```python
n_producers = [126, 50, 49, 34, 26]
n_producers_cols = ['SP', 'RJ', 'RS', 'PE', 'PR']

percents = [49.56, 12.96, 11.63, 9.66, 4.89]
percents_cols = ['SP', 'PE', 'RJ', 'PR', 'RS']

data = [[n_producers_cols[i], n_producers[i]] for i in range(len(n_producers_cols))]
df_prod = pd.DataFrame(
    np.array(data),
    columns=['Estado', 'Produtores'],
)
df_prod['Produtores'] = pd.to_numeric(df_prod['Produtores'])

data = [[percents_cols[i], percents[i]] for i in range(len(percents_cols))]
df_exp = pd.DataFrame(
    np.array(data),
    columns=['Estado', 'Porcentagem'],
)
df_exp['Porcentagem'] = pd.to_numeric(df_exp['Porcentagem'])

sns.barplot(x="Produtores", y="Estado", palette="Blues_d", data=df_prod)
plt.show()

sns.barplot(x="Porcentagem", y="Estado", palette="Blues_d", data=df_exp)
plt.show()
```
